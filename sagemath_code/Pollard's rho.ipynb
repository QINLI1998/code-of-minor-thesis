```python
# ============================================================================
# 70-bit 有限域上的 Pollard's rho 基准测试
# ============================================================================
import time
from sage.groups.generic import discrete_log

# 生成 60-bit 素数 p (2^59 ~ 2^60)
p = random_prime(2^99, lbound=2^98)
print(f"素数 p = {p} (比特长度: {p.nbits()})")
F = GF(p)

# 构造普通安全曲线（要求最大素因子至少 30 比特，保证破解时间在数秒量级）
while True:
    a = F.random_element()
    b = F.random_element()
    try:
        E = EllipticCurve(F, [a, b])
        order = E.order()
        if order == p or E.is_supersingular():
            continue
        factors = order.factor()
        max_prime = max([f[0] for f in factors])
        if max_prime.nbits() >= 30:
            break
    except:
        continue

print(f"曲线: y^2 = x^3 + {a}x + {b}")
print(f"阶: {order} = {order.factor()}")

big_prime = max_prime
P = (order // big_prime) * E.gens()[0]
print(f"生成元阶: {P.order()} (比特长度: {P.order().nbits()})")

k_true = randint(1, big_prime - 1)
Q = k_true * P

start = time.time()
k_found = discrete_log(Q, P, operation='+', algorithm='rho')
elapsed = time.time() - start

print(f"预期私钥: {k_true}")
print(f"求解私钥: {k_found}")
print(f"Pollard's rho 破解时间: {elapsed:.4f} 秒")
```
