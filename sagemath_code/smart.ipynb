```python
# ============================================================================
# Smart 攻击 - 使用已知的异常曲线（PoliCTF 2012）
# ============================================================================
import time

def SmartAttack(P, Q, p):
    """
    Smart 攻击：针对异常曲线（#E(Fp)=p）的线性时间 ECDLP 求解算法。
    """
    E = P.curve()
    Eqp = EllipticCurve(Qp(p, 2), [ZZ(t) + randint(0, p)*p for t in E.a_invariants()])
    
    P_Qps = Eqp.lift_x(ZZ(P.xy()[0]), all=True)
    for P_Qp in P_Qps:
        if GF(p)(P_Qp.xy()[1]) == P.xy()[1]:
            break
            
    Q_Qps = Eqp.lift_x(ZZ(Q.xy()[0]), all=True)
    for Q_Qp in Q_Qps:
        if GF(p)(Q_Qp.xy()[1]) == Q.xy()[1]:
            break
    
    p_times_P = p * P_Qp
    p_times_Q = p * Q_Qp
    x_P, y_P = p_times_P.xy()
    x_Q, y_Q = p_times_Q.xy()
    
    phi_P = -(x_P / y_P)
    phi_Q = -(x_Q / y_Q)
    
    k = phi_Q / phi_P
    return ZZ(k)

# 使用 PoliCTF 2012 的异常曲线参数
print("="*60)
print("Smart 攻击 - 使用已知异常曲线（PoliCTF 2012）")
print("="*60)

# 定义曲线参数
p = 730750818665451459112596905638433048232067471723
F = GF(p)
a = 425706413842211054102700238164133538302169176474
b = 203362936548826936673264444982866339953265530166
E = EllipticCurve(F, [a, b])

# 定义点
P = E(125270202464411072778547771568975423382990845665,
       440970603958123875213441435758390311809187352362)
Q = E(296939092187233862778999244256460019221379646447,
       650021996391906816753000782344884033217784284632)

print(f"有限域 p = {p}")
print(f"曲线阶: {E.order()}")
print(f"验证 #E == p: {E.order() == p}\n")

start = time.time()
k_found = SmartAttack(P, Q, p)
elapsed = time.time() - start

print(f"求解私钥: {k_found}")
print(f"Smart 攻击破解时间: {elapsed:.6f} 秒")
```
