```python
import pickle
import time

# 配置
bit_lengths = [128, 160, 192, 224]
NUM_CURVES = 500 

results = {b: [] for b in bit_lengths}
print(f"开始生成数据...")
start_all = time.time()

for b in bit_lengths:
    p = random_prime(2**b - 1, lbound=2**(b-1))
    F = GF(p)
    count = 0
    print(f"正在处理 {b} 位域...")
    while count < NUM_CURVES:
        a, b_val = F.random_element(), F.random_element()
        if (4*a**3 + 27*b_val**2) == 0: continue
        try:
            E = EllipticCurve(F, [a, b_val])
            N = E.order()
            factors = factor(N)
            n = factors[-1][0]
            results[b].append({
                'p': p, 'N': N, 'n': n, 'h': int(N//n), 'n_bits': n.nbits()
            })
            count += 1
        except: continue

# 保存数据到本地文件
with open('ecc_data_4.pkl', 'wb') as f:
    pickle.dump(results, f)

print(f"数据已保存至 ecc_data.pkl。总耗时: {time.time()-start_all:.2f}s")
# 1. 定义文件名
file1 = 'ecc_data_3.pkl'
file2 = 'ecc_data_4.pkl'
output_file = 'ecc_data_merged.pkl'

# 2. 读取第一个包
with open(file1, 'rb') as f:
    data_total = pickle.load(f)

# 3. 读取第二个包并合并
with open(file2, 'rb') as f:
    data_new = pickle.load(f)

# 4. 遍历合并列表
for bit_length, curves in data_new.items():
    if bit_length in data_total:
        # 如果键已存在，把新生成的曲线列表加到旧列表后面
        data_total[bit_length].extend(curves)
    else:
        # 如果是新位长，直接赋值
        data_total[bit_length] = curves

# 5. 打印统计信息，确认合并成功
print("合并完成！当前数据统计：")
for b in sorted(data_total.keys()):
    print(f"{b}-bit 域下的曲线总数: {len(data_total[b])}")

# 6. 保存合并后的数据
with open(output_file, 'wb') as f:
    pickle.dump(data_total, f)

print(f"\n合并后的数据已保存至: {output_file}")
```
