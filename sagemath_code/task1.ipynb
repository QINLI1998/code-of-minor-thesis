```python
import pickle
import matplotlib.pyplot as plt
from sage.all import sqrt # 这里的代码仍建议在 sage 环境下运行以支持大数开方

with open('ecc_data_2.pkl', 'rb') as f:
    results = pickle.load(f)

data_256 = results[256]
p_256 = data_256[0]['p']
center = p_256 + 1
# 关键修复：计算偏差值
deviations = [float(d['N'] - center) for d in data_256]
bound = float(2 * sqrt(p_256))

plt.figure(figsize=(8, 5))
weights_10 = [5] * len(deviations) 
plt.hist(deviations, bins=20, weights=weights_10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(-bound, color='red', linestyle='--', label='Hasse Lower')
plt.axvline(bound, color='green', linestyle='--', label='Hasse Upper')
plt.title('Figure 4-3: Distribution of Curve Orders over F_256(Deviation from p+1)')
plt.xlabel('Deviation Value')
plt.ylabel('Frequency')
plt.legend()
plt.savefig('fig4-3_hasse.png', dpi=300)
print("图 4-3 已生成。")
```
