```python
# ==========================================
# 4.1 单条椭圆曲线的计算与可视化 
# ==========================================

# 1. 示例曲线选择与基础属性验证
print("=== 步骤 1: 曲线定义与参数验证 ===")
p = 101
F = GF(p)
a = -1
b = 1

# 定义椭圆曲线 E: y^2 = x^3 - x + 1 上的有限域 F_101
E = EllipticCurve(F, [a, b])
print(f"定义的椭圆曲线: {E}")

# 计算阶数
order = E.order()
print(f"曲线的阶 #E = {order}")

# 计算判别式并验证非奇异性
delta = E.discriminant()
print(f"判别式 Delta = {delta}")
if delta != 0:
    print("结论: Delta != 0, 满足非奇异条件，该曲线有效。")
else:
    print("结论: Delta == 0, 这是一个奇异曲线！")
print("-" * 40)


# 2. 离散点集可视化
print("=== 步骤 2: 生成离散点分布图 ===")
# 获取曲线上所有点，过滤掉无穷远点
points = E.points()
# 【关键修复】：加上 int() 将有限域元素转换为普通整数，避免绘图报错
points_xy = [(int(P[0]), int(P[1])) for P in points if not P.is_zero()]

# 绘制散点图
# 准备要显示的文本信息
curve = "E:y^2 = x^3 + 100*x + 1"
info_text = f"Order(#E)={order}, Delta={delta}"

# 1. 修改第一张图：将信息放入标题
print("绘制图 4-1...")
fig1 = list_plot(points_xy, size=30, color='blue', 
                 # 在标题中动态加入阶和判别式
                 title=f"Discrete Points of {curve} over E(F_101)\n({info_text})",
                 axes_labels=['x', 'y'], figsize=6)

fig1.save("fig4-1_discrete_points.png")
print("图 4-1 已生成并保存为 'fig4-1_discrete_points.png'")


# 3. 标量乘法轨迹演示
print("=== 步骤 3: 生成标量乘法轨迹图 ===")
# 设定生成元 P = (0, 1)
P = E(0, 1)

# 计算 P, 2P, 3P, ..., 20P
trajectory_points = [k * P for k in range(1, 21)]

# 初始化一个空图层，先将所有背景点用浅灰色画出，作为底图
# 2. 修改第二张图：将信息作为图内标签放置
fig2 = list_plot(points_xy, size=20, color='lightgray', figsize=7,
                 title=f"Scalar Multiplication Trajectory of {curve} over E(F_101)\n({info_text})",
                 axes_labels=['x', 'y'])

# 在图的左上角添加一块背景说明文字 (坐标根据 p=101 适当调整，比如选在 x=5, y=95 处)
# horizontal_alignment='left' 确保文字向右延伸
# fig2 += text(info_text, (5, 95), fontsize=10, color='darkred', 
#              background_color='white', bbox=dict(boxstyle='round', fc='w', ec='0.5', alpha=0.7),
#              horizontal_alignment='left', zorder=10)

# 【关键修复】：将轨迹坐标也转为标准整数
valid_trajectory = [(int(pt[0]), int(pt[1])) for pt in trajectory_points if not pt.is_zero()]

# 使用循环绘制点、箭头和标签（加入颜色渐变表示跳跃顺序）
for i in range(len(valid_trajectory) - 1):
    pt1 = valid_trajectory[i]
    pt2 = valid_trajectory[i+1]
    
    # 根据跳跃步数生成渐变色 (从红色过渡到紫色)
    color_hue = hue(i / 25.0) 
    
    # 【关键修复】：给 pt1 套上中括号 [pt1]，将其包装为列表传入 point2d
    fig2 += point2d([pt1], size=50, color=color_hue, zorder=5)
    
    # 添加文字标签 (例如: 1P, 2P...)，因为已经是普通 int，+3 只是纯粹的视觉上移
    fig2 += text(f"{i+1}P", (pt1[0], pt1[1] + 3), fontsize=8, color='black', zorder=6)
    
    # 绘制指向下一个点的箭头
    fig2 += arrow2d(pt1, pt2, color=color_hue, width=0.5, arrowsize=2, zorder=4)

# 单独处理最后一个点 (20P)
last_pt = valid_trajectory[-1]
last_color = hue(len(valid_trajectory) / 25.0)
fig2 += point2d([last_pt], size=50, color=last_color, zorder=5)
fig2 += text(f"{len(valid_trajectory)}P", (last_pt[0], last_pt[1] + 3), fontsize=8, color='black', zorder=6)

# 保存图表供 LaTeX 引用
fig2.save("fig4-2_scalar_mul_trajectory.png")
print("图 4-2 已生成并保存为 'fig4-2_scalar_mul_trajectory.png'")
```
